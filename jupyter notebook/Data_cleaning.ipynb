```python
import pandas as pd
df_customers = pd.read_csv(r"D:\python project files\bank profitbility\data\Raw\customers.csv")
df_accounts = pd.read_csv(r"D:\python project files\bank profitbility\data\Raw\accounts.csv") 
df_transactions = pd.read_csv(r"D:\python project files\bank profitbility\data\Raw\transactions.csv")
df_payments = pd.read_csv(r"D:\python project files\bank profitbility\data\Raw\payments.csv")
```

```python
#drop duplicates

df_customers = df_customers.drop_duplicates()
df_accounts = df_accounts.drop_duplicates()
df_transactions = df_transactions.drop_duplicates()
df_payments = df_payments.drop_duplicates()
```

```python
#drop null values
df_customers = df_customers.dropna()
df_acounts = df_accounts.dropna()
df_transactions = df_transactions.dropna()
df_payments = df_payments.dropna()
```

```python
# Example type fixes
if "customer_since" in df_customers.columns:
    df_customers["customer_since"] = pd.to_datetime(df_customers["customer_since"], errors="coerce")
if "open_date" in df_accounts.columns:
    df_accounts["open_date"] = pd.to_datetime(df_accounts["open_date"], errors="coerce")
```

```python
df_customers.to_csv(r"D:\python project files\bank profitbility\data\Cleaned data\customers.csv")
df_accounts.to_csv(r"D:\python project files\bank profitbility\data\Cleaned data\accounts.csv")
df_transactions.to_csv(r"D:\python project files\bank profitbility\data\Cleaned data\transactions.csv")
df_payments.to_csv(r"D:\python project files\bank profitbility\data\Cleaned data\payments.csv")
```
